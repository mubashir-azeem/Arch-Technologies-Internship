```python
import os

base_path = "/content/data/000"
images = os.listdir(base_path)

images = images[:10]

print(images)

#add the dataset to directory and displayed the name of some images.
```

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt

#imported the libraries.
```

```python
img_path = base_path + "/" + images[0]

img = cv2.imread(img_path)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')

#selected an image and read it thorugh cv2.imread and displayed it.
```

```python
text = pytesseract.image_to_string(img_rgb)

print("Extracted Text:\n")
print(text)

#extracted the text from the image using the pytesseact.
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

plt.imshow(thresh, cmap='gray')
plt.axis('off')

#for better output converted the image on greyscale colors and processed it.
```

```python
# Extracted text from processed image
text2 = pytesseract.image_to_string(thresh)

print("Improved OCR Text:\n")
print(text2)
```

```python
results = []

for img_name in images:

    img_path = base_path + "/" + img_name

    img = cv2.imread(img_path)

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

    text = pytesseract.image_to_string(thresh)

    results.append((img_name, text))

    print("="*50)
    print("Image:", img_name)
    print("Extracted Text:\n", text)


#did all that for images from the dataset.
```
